```python
# インポート
import os
import sys
import json
import numpy as np
import pandas as pd
from itertools import chain
```

```python
# パスの取得
def loadingJson(dirpath, f):

    fpath = dirpath + '/' + f
    fj = open(fpath,'r')
    json_data = json.load(fj)
    fj.close()
    
    return json_data

# データファイルをパース
def output(data):
    
    #utterances = []
    #responses = []
    dialogue = []
    for i in range(len(data['turns'])):
        dialogue.append(data['turns'][i]['utterance'])
        """
        if data['turns'][i]['speaker'] == "S":
            utterances.append(data['turns'][i]['utterance'])
        elif data['turns'][i]['speaker'] == "U":
            responses.append(data['turns'][i]['utterance'])
        else:
            continue
        """
            
    return dialogue[:-1]
```

```python
# ディレクトリのファイル一覧を取得して、対話データを作成

# 分割の入れ子リストを作成
def split_list(l, n):
    """
    リストをサブリストに分割する
    :param l: リスト
    :param n: サブリストの要素数
    :return: 
    """
    for idx in range(0, len(l), n):
        yield l[idx:idx + n]

branch = os.walk("online_chat_data")

sequence = []
for dirpath, dirs, files in branch:
    for f in files:
        json_data = loadingJson(dirpath, f)
        sequence.append(output(json_data))

flatten = list(chain.from_iterable(sequence))
nest_utterance = list(split_list(flatten, 1))
nest_interaction = list(split_list(nest_utterance, 2))

#print(files)
print(nest_interaction)
```
